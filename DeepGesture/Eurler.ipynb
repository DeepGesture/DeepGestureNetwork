```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Sample Data (Sine wave with noise)
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # Time variable
f = 5  # Frequency of the sine wave
amplitude = 1
offset = 0.5
# Create a sample periodic signal
signal = amplitude * np.sin(2 * np.pi * f * t) + offset + 0.1 * np.random.normal(size=t.shape)
```

```python
# Step 2: Compute the Fourier Transform
fourier_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1/fs)
```

```python
# Step 3: Extract Amplitude and Phase
# Only take the positive frequencies for visualization
positive_frequencies = frequencies[:len(frequencies)//2]
amplitude_spectrum = np.abs(fourier_transform)[:len(fourier_transform)//2]
phase_spectrum = np.angle(fourier_transform)[:len(fourier_transform)//2]

```

```python

# Find the dominant frequency
dominant_freq_index = np.argmax(amplitude_spectrum)
dominant_freq = positive_frequencies[dominant_freq_index]
dominant_amplitude = amplitude_spectrum[dominant_freq_index]
dominant_phase = phase_spectrum[dominant_freq_index]
```

```python

# Step 4: Convert to Circular Frequency Domain
circular_frequencies = 2 * np.pi * positive_frequencies
```

```python
# Step 5: Visualize the Results
plt.figure(figsize=(12, 10))

# Original Signal
plt.subplot(4, 1, 1)
plt.plot(t, signal, label='Original Signal', color='blue')
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

# Amplitude Spectrum
plt.subplot(4, 1, 2)
plt.stem(positive_frequencies, amplitude_spectrum, 'r', markerfmt=" ", basefmt=" ")
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.grid(True)

# Phase Spectrum
plt.subplot(4, 1, 3)
plt.stem(positive_frequencies, phase_spectrum, 'g', markerfmt=" ", basefmt=" ")
plt.title('Phase Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [radians]')
plt.grid(True)

# Circular Frequency Spectrum
plt.subplot(4, 1, 4)
plt.stem(circular_frequencies, amplitude_spectrum, 'm', markerfmt=" ", basefmt=" ")
plt.title('Amplitude Spectrum in Circular Frequency Domain')
plt.xlabel('Circular Frequency [rad/s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

# Output the dominant frequency, amplitude, and phase
print(f'Dominant Frequency: {dominant_freq:.2f} Hz')
print(f'Dominant Amplitude: {dominant_amplitude:.2f}')
print(f'Dominant Phase: {dominant_phase:.2f} radians')
print(f'Dominant Circular Frequency: {2 * np.pi * dominant_freq:.2f} rad/s')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create values for x
x = np.linspace(0, 2 * np.pi, 100)

# Calculate the real and imaginary parts of e^ix
cos_x = np.cos(x)  # Real part: cos(x)
sin_x = np.sin(x)  # Imaginary part: sin(x)

# Create the figure and the 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D spiral curve of e^ix
ax.plot(cos_x, sin_x, x, label=r'$e^{ix} = \cos(x) + i\sin(x)$')

# Label axes
ax.set_xlabel('Real (cos(x))')
ax.set_ylabel('Imaginary (sin(x))')
ax.set_zlabel('Angle (x)')

# Add title and legend
ax.set_title(r'3D Visualization of $e^{ix}$')
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create values for x
x = np.linspace(0, 2 * np.pi, 400)

# Calculate cos(x) and sin(x)
cos_x = np.cos(x)
sin_x = np.sin(x)

# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot the cosine wave (Real part of e^ix)
axs[0].plot(x, cos_x, label=r'$\cos(x)$', color='b')
axs[0].set_title('Real Part (Cosine)')
axs[0].set_xlabel('x')
axs[0].set_ylabel(r'$\cos(x)$')
axs[0].grid(True)

# Plot the sine wave (Imaginary part of e^ix)
axs[1].plot(x, sin_x, label=r'$\sin(x)$', color='r')
axs[1].set_title('Imaginary Part (Sine)')
axs[1].set_xlabel('x')
axs[1].set_ylabel(r'$\sin(x)$')
axs[1].grid(True)

# Plot the circular motion of e^ix
axs[2].plot(cos_x, sin_x, label=r'$e^{ix} = \cos(x) + i\sin(x)$', color='g')
axs[2].set_title(r'Circular Motion ($e^{ix}$)')
axs[2].set_xlabel(r'Real ($\cos(x)$)')
axs[2].set_ylabel(r'Imaginary ($\sin(x)$)')
axs[2].set_aspect('equal')
axs[2].grid(True)

# Add legends to each subplot
for ax in axs:
    ax.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
